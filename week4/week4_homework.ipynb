## 2.1. Kādā situācijā lietot kuru pieeju?

**Klasifikāciju** lieto tad, ja mērķa mainīgais ir kategorija vai klase.  
Piemēram, veselības aprūpē klasifikācijas modelis var paredzēt, vai pacients ieradīsies uz vizīti vai arī to kavēs. Tā kā iespējamais rezultāts ir “jā” vai “nē”, šeit ir piemērota klasifikācija.

**Regresiju** lieto tad, ja mērķa mainīgais ir nepārtraukta skaitliska vērtība.  
Piemēram, nekustamo īpašumu jomā ar regresijas palīdzību var prognozēt dzīvokļa cenu, izmantojot tā platību, atrašanās vietu, stāvu un citus parametrus. Šajā situācijā rezultāts ir skaitlis, tāpēc regresija ir pareizā pieeja.

**Klasterizāciju** lieto tad, ja datiem nav iepriekš definēta mērķa mainīgā un mēs vēlamies atrast dabiskas grupas datos.  
Piemēram, telekomunikāciju uzņēmums var segmentēt klientus pēc viņu lietošanas paradumiem, lai saprastu, kādi klientu tipi eksistē un kā viņiem pielāgot pakalpojumus.

**Pipeline** ir vajadzīgs tāpēc, ka tas apvieno priekšapstrādi un modeli vienā secīgā, atkārtojamā procesā.  
Tas palīdz nodrošināt, ka vienas un tās pašas transformācijas tiek piemērotas gan treniņdatiem, gan testēšanas datiem, tādējādi samazinot kļūdu risku.

**Cross-validation** ir nepieciešams tāpēc, ka viens train/test sadalījums var dot nejauši labāku vai sliktāku rezultātu.  
Izmantojot vairākus foldus, mēs iegūstam stabilāku priekšstatu par modeļa kvalitāti un tā spēju vispārināt jaunus datus.
