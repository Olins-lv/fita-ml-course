```python
# 5. uzdevums. Bāzes modelis (Baseline)

import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import json
import ast

# Izveido kopiju, lai tieši nemodificētu sākotnējo DataFrame
df = df_final.copy()

# 1. Neatbilstošo kolonnu nometšana
columns_to_drop = [
    'belongs_to_collection', 'homepage', 'backdrop_path', 'id', 'imdb_id',
    'original_title', 'overview', 'poster_path', 'tagline', 'title',
    'spoken_languages', 'production_companies', 'credits', 'keywords',
    'adult', 'video'
]
df = df.drop(columns=columns_to_drop, errors='ignore')

# 2. Nederīgu budžeta un ieņēmumu vērtību filtrēšana
df = df[(df['revenue'] > 0) & (df['budget'] > 1)]

# 3. JSON kolonnu apstrāde (`genres`, `production_countries`, `origin_country`)
def extract_first_name(data_str):
    try:
        if isinstance(data_str, str):
            items = ast.literal_eval(data_str)
            if items and len(items) > 0 and isinstance(items, list) and 'name' in items[0]:
                return items[0]['name']
        return np.nan
    except (ValueError, SyntaxError, TypeError):
        return np.nan

def extract_first_country_from_list_string(list_str):
    try:
        if isinstance(list_str, str):
            items = ast.literal_eval(list_str)
            if items and len(items) > 0:
                return items[0] # Pieņemot, ka elementi ir valstu kodi, piemēram, 'US' (Assuming items are country codes like 'US')
        return np.nan
    except (ValueError, SyntaxError):
        return np.nan

columns_to_extract = {
    'genres': 'primary_genre',
    'production_countries': 'primary_production_country',
    'origin_country': 'primary_origin_country'
}

for original_col, new_col in columns_to_extract.items():
    if original_col in df.columns:
        if original_col == 'origin_country':
            df[new_col] = df[original_col].apply(extract_first_country_from_list_string)
        else:
            df[new_col] = df[original_col].apply(extract_first_name)

cols_to_drop_after_extraction = [col for col in columns_to_extract.keys() if col in df.columns]
if cols_to_drop_after_extraction:
    df = df.drop(columns=cols_to_drop_after_extraction)

# 4. `release_date` apstrāde
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year
df = df.drop(columns=['release_date'])

# 5. Logaritmiskā transformācija `budget` un `revenue`
budget_median = df['budget'].median()
df['budget'] = df['budget'].fillna(budget_median)
df['budget'] = np.log1p(df['budget'])
df['revenue'] = np.log1p(df['revenue'])

# 6. X un y definēšana
X = df.drop('revenue', axis=1)
y = df['revenue']


# 7. Bāzes modeļa (Baseline Model) Pipeline veidošana un novērtēšana
# Definē skaitliskās un kategoriskās iezīmes priekšapstrādei
numerical_features = ['budget', 'popularity', 'runtime', 'vote_average', 'vote_count']

# Inicializē kategoriskās iezīmes ar tām, kuras vienmēr ir paredzētas vai ir veiksmīgi izveidotas
categorical_features = ['original_language', 'status', 'release_year']

# Dinamiski pievieno 'primary_genre', 'primary_production_country' un 'primary_origin_country'
# tikai tad, ja tās pastāv X DataFrame

if 'primary_genre' in X.columns:
    categorical_features.append('primary_genre')
if 'primary_production_country' in X.columns:
    categorical_features.append('primary_production_country')
if 'primary_origin_country' in X.columns:
    categorical_features.append('primary_origin_country')

# Izveido priekšapstrādes pipelines skaitliskajām un kategoriskajām iezīmēm
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler()) # Standartizē skaitliskās iezīmes (Standardize numerical features)
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')), # Aizpilda trūkstošās vērtības ar konstantu (Impute missing values with a constant)
    ('onehot', OneHotEncoder(handle_unknown='ignore')) # Veic One-Hot kodēšanu kategoriskajām iezīmēm (Perform One-Hot encoding for categorical features)
])

# Izveido preprocessor, izmantojot ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='drop' # Nomet visas kolonnas, kas nav norādītas (Drop any columns not specified)
)

# Izveido pilnu pipeline ar preprocessor un lineārās regresijas modeli

baseline_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression()) # Izmanto lineāro regresiju kā bāzes modeli (Use Linear Regression as the baseline model)
])

# Novērtē bāzes modeli, izmantojot krustenisko validāciju
# Evaluate the baseline model using cross-validation
r2_scores = cross_val_score(baseline_pipeline, X, y, cv=5, scoring='r2', n_jobs=-1)

print(f"Bāzes modeļa R-squared rezultāti: {r2_scores}")
print(f"Vidējais R-squared: {r2_scores.mean():.4f} +/- {r2_scores.std():.4f}")
```

## 6.uzdevums. Feature Engineering un datu sagatavošana


### 6.1. Kategorisko kolonnu kodēšana

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
import pandas as pd

# Izveidojam datu kopijas versiju feature inženierijai
df_engineered = df.copy()

# Identificē kategoriskās un skaitliskās kolonnas
numerical_cols = ['budget', 'popularity', 'runtime', 'vote_average', 'vote_count']
categorical_cols = ['original_language', 'status', 'primary_genre', 'primary_production_country', 'primary_origin_country']

# Pārbauda un grupē kolonnas ar pārāk daudz unikālām vērtībām (>20)
print("Unikālo vērtību skaits PIRMS grupēšanas:")
for col in categorical_cols:
    if col in df_engineered.columns:
        n_unique = df_engineered[col].nunique()
        print(f"{col}: {n_unique}")

        # Ja vērtību ir vairāk par 20, atstājam top 15 un pārējās kļūst par 'Other'
        if n_unique > 20:
            top_categories = df_engineered[col].value_counts().nlargest(15).index
            df_engineered[col] = df_engineered[col].apply(lambda x: x if x in top_categories else 'Other')

print("\nUnikālo vērtību skaits PĒC grupēšanas:")
for col in categorical_cols:
    if col in df_engineered.columns:
        print(f"{col}: {df_engineered[col].nunique()}")

# Filtrejam reāli eksistējošās kategoriskās kolonnas no saraksta
existing_cat_cols = [c for c in categorical_cols if c in df_engineered.columns]

# Izveido ColumnTransformer ar atbilstošiem transformeriem
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), existing_cat_cols)
    ],
    remainder='passthrough' # vai 'drop' atkarībā no vajadzības
)

print("\nColumnTransformer (preprocessor) ir veiksmīgi izveidots!")
```

### 6.2. Trūkstošo vērtību apstrāde

**Lēmumu pamatojums (balstoties uz 4.3. stratēģiju):**
1. **Kolonnu noņemšana (>50% tukšas):** Kolonnas kā `belongs_to_collection` un `homepage` jau tika izmestas bāzes modeļa fāzē, taču kodā mēs ieliksim dinamisku pārbaudi, lai izmestu jebkuru kolonnu, kas ir vairāk nekā 50% tukša.
2. **Skaitliskās kolonnas:** Izmantosim `SimpleImputer(strategy='median')`. Mediāna ir ieteicama asimetriskiem sadalījumiem (kā budžets un ieņēmumi), jo to neietekmē ekstrēmi izlēcēji (outliers).
3. **Kategoriskās kolonnas:** Izmantosim `SimpleImputer(strategy='most_frequent')`, lai aizpildītu tukšās vērtības ar vispopulārāko kategoriju.

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# 1. Pārbaudām, vai ir atlikušas kolonnas ar >50% trūkstošo vērtību
missing_percentages = df_engineered.isnull().sum() / len(df_engineered)
cols_to_drop = missing_percentages[missing_percentages > 0.5].index.tolist()

if cols_to_drop:
    print(f"Dzēšam kolonnas ar >50% trūkstošajām vērtībām: {cols_to_drop}")
    df_engineered = df_engineered.drop(columns=cols_to_drop)
else:
    print("Nav atlikušu kolonnu ar >50% trūkstošajām vērtībām (tās jau tika izmestas iepriekš).")

# Atjauninām sarakstus pēc iespējamās kolonnu dzēšanas
numerical_cols = [col for col in numerical_cols if col in df_engineered.columns]
existing_cat_cols = [col for col in existing_cat_cols if col in df_engineered.columns]

# 2. Skaitliskajām kolonnām - SimpleImputer(strategy='median') + StandardScaler
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# 3. Kategoriskajām kolonnām - SimpleImputer(strategy='most_frequent') + OneHotEncoder
cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# 4. Atjauninām ColumnTransformer ar jaunajiem pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, numerical_cols),
        ('cat', cat_pipeline, existing_cat_cols)
    ],
    remainder='passthrough' # Pārējās kolonnas atstājam neskartas (ja tādas ir)
)

print("\n'preprocessor' ir veiksmīgi atjaunināts ar trūkstošo vērtību apstrādes (Imputer) soļiem!")
```

### 6.3. Jaunu pazīmju izveide (Feature Creation)

```python
import pandas as pd

# 1. Attiecība (ratio) starp divām kolonnām: Budžets uz vienu filmas minūti
# Pievienojam +1 runtime, lai izvairītos no dalīšanas ar nulli anomāliju gadījumā
df_engineered['budget_per_minute'] = df_engineered['budget'] / (df_engineered['runtime'] + 1)

# 2. Binning (intervālu grupēšana): Filmas garuma (runtime) kategorizēšana
bins = [0, 90, 120, 180, 1000]
labels = ['Short', 'Average', 'Long', 'Epic']
df_engineered['runtime_category'] = pd.cut(df_engineered['runtime'], bins=bins, labels=labels)

# 3. Papildus skaitliskā pazīme: Popularitātes un reitinga mijiedarbība
df_engineered['popularity_vote_score'] = df_engineered['popularity'] * df_engineered['vote_average']

# Pārbaudām korelāciju ar target kolonnu ('revenue')
# Kategoriskajai kolonnai korelāciju nerēķinām tieši, bet varam apskatīties jaunās skaitliskās pazīmes
new_numeric_features = ['budget_per_minute', 'popularity_vote_score', 'revenue']
correlation_matrix = df_engineered[new_numeric_features].corr()

print("Korelācija ar target kolonnu (revenue):")
display(correlation_matrix[['revenue']])

# Pievienojam jaunās kolonnas mūsu apstrādes sarakstiem, lai ColumnTransformer tās paķertu vēlāk
numerical_cols.extend(['budget_per_minute', 'popularity_vote_score'])
existing_cat_cols.append('runtime_category')

print("\nJaunās pazīmes veiksmīgi izveidotas un pievienotas apstrādes sarakstiem!")
```

## 7.uzdevums. Modeļu salīdzināšana un hiperparametru optimizācija

### 7.1. Salīdzinu vismaz 3 modeļus

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
import time

# Atjauninām X un y ar jaunajām pazīmēm un apstrādātajiem datiem no 6. uzdevuma
X_eng = df_engineered.drop('revenue', axis=1)
y_eng = df_engineered['revenue']

# Definē modeļu sarakstu
models = {
    'Ridge Regression': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', Ridge())
    ]),
    'Random Forest': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', RandomForestRegressor(random_state=42, n_jobs=-1))
    ]),
    'Gradient Boosting': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', GradientBoostingRegressor(random_state=42))
    ])
}

# Izpildi cross_val_score katram modelim un saglabā rezultātus
model_results = {}

print("Sākas modeļu novērtēšana (tas var aizņemt nedaudz laika)...\n")

for name, pipe in models.items():
    start_time = time.time()
    print(f"Apmāca un vērtē {name}...")

    # Izmantojam R-squared metriku, gluži kā bāzes modelim
    scores = cross_val_score(pipe, X_eng, y_eng, cv=5, scoring='r2', n_jobs=-1)

    model_results[name] = scores
    elapsed_time = time.time() - start_time

    print(f"Vajadzīgais laiks: {elapsed_time:.2f} sekundes")
    print(f"Rezultāts ({name}): {scores.mean():.4f} ± {scores.std():.4f}\n")

print("Visu modeļu novērtēšana pabeigta!")
```

### 7.2. Hiperparametru optimizācija ar GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
import time

# 1. Izvēlamies labāko modeli no iepriekšējā soļa (Random Forest Pipeline)
best_pipe = models['Random Forest']

# 2. Definējam parametru režģi
# Piezīme: Tā kā izmantojam Pipeline, parametriem jāpievieno prefikss 'regressor__'
param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [10, 20, None],
    'regressor__min_samples_split': [2, 5]
}

print("Sākas GridSearchCV hiperparametru optimizācija (tas var aizņemt vairākas minūtes)...\n")

# 3. Izveidojam un apmācam GridSearchCV
# scoring='r2', jo tas ir regresijas uzdevums
grid_search = GridSearchCV(best_pipe, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=1)

start_time = time.time()
grid_search.fit(X_eng, y_eng)
elapsed_time = time.time() - start_time

# 4. Izdrukājam labākos parametrus un rezultātu
print(f"\nOptimizācija pabeigta {elapsed_time:.2f} sekundēs.")
print(f"Labākie parametri: {grid_search.best_params_}")
print(f"Labākais CV R-squared rezultāts: {grid_search.best_score_:.4f}")
```

### 7.3. Modeļu salīdzinājuma tabula un secinājumi

| Modelis | CV Score (vidējais R²) | CV Score (std) | Piezīmes |
| :--- | :--- | :--- | :--- |
| Bāzes (Lineārā regresija) | -0.6007 | 1.9189 | Sākotnējais rezultāts (stipra pārkārtošanās) |
| Ridge Regresija | 0.2443 | 0.2822 | Ievērojams uzlabojums no bāzes, labāka regularizācija |
| Gradient Boosting | 0.4425 | 0.0292 | Spēcīgs modelis nelineārām sakarībām |
| Random Forest | 0.4536 | 0.0271 | Labākais no izvērtētajiem 3 modeļiem |
| GridSearchCV (Random Forest) | 0.4566 | - | Optimizēts (max_depth=None, n_estimators=200, min_samples_split=2) |

**Secinājumi:**
Labākais modelis izrādījās `Random Forest` pēc GridSearchCV hiperparametru optimizācijas. Uzlabojums no bāzes modeļa ir milzīgs – kamēr lineārās regresijas bāzes modelis nespēja pielāgoties datiem un uzrādīja negatīvu R² (un ļoti augstu dispersiju/std), koku bāzētie algoritmi (Kā Random Forest un Gradient Boosting) uzrādīja R² virs 0.45. Random Forest uzvarēja, jo koku ansambļu modeļi ļoti labi tiek galā ar nelineārām sakarībām, izlēcējiem un sarežģītām kategorisko/skaitlisko iezīmju mijiedarbībām (kā filmas budžets pret ieņēmumiem un popularitāti), kas lineāriem modeļiem sagādā problēmas bez ļoti specifiskas iepriekšējas datu transformācijas.

## 8.uzdevums. Modeļa interpretācija un rezultātu vizualizācija



### 8.1. Feature Importance

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Iegūstam labāko apmācīto modeli no GridSearchCV (Pipeline)
best_model = grid_search.best_estimator_

# 2. Iegūstam feature importances no Random Forest regressor soļa
importances = best_model.named_steps['regressor'].feature_importances_

# 3. Iegūstam ģenerētos pazīmju nosaukumus no preprocessor (ColumnTransformer)
feature_names = best_model.named_steps['preprocessor'].get_feature_names_out()

# 4. Izveidojam DataFrame ērtākai šķirošanai un attēlošanai
feat_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# Sakārtojam dilstošā secībā un ņemam top 10
top_10_features = feat_imp_df.sort_values(by='Importance', ascending=False).head(10)

# 5. Vizualizējam ar Seaborn horizontālo barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=top_10_features, palette='viridis', hue='Feature', legend=False)
plt.title('Top 10 Svarīgākās Pazīmes (Random Forest Modelis)')
plt.xlabel('Svarīgums (Feature Importance)')
plt.ylabel('Pazīmes (Features)')
plt.tight_layout()
plt.show()
```

### 8.2. Galvenā novērtēšanas vizualizācija (Regresija)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_predict

# Tā kā iepriekš izmantojām CV, nevis klasisku train_test_split,
# ģenerējam out-of-fold prognozes reālistiskam novērtējumam
print("Ģenerē prognozes (tas var nedaudz ieilgt)...")
y_pred = cross_val_predict(best_model, X_eng, y_eng, cv=5, n_jobs=-1)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Faktiskās vs. Prognozētās vērtības (Scatter plot)
axes[0].scatter(y_eng, y_pred, alpha=0.3, color='teal')
axes[0].plot([y_eng.min(), y_eng.max()], [y_eng.min(), y_eng.max()], 'r--', lw=2)
axes[0].set_title('Faktiskie vs. Prognozētie Ieņēmumi (Log skala)')
axes[0].set_xlabel('Faktiskā vērtība (Real Revenue)')
axes[0].set_ylabel('Prognozētā vērtība (Predicted Revenue)')
axes[0].grid(True, linestyle='--', alpha=0.7)

# 2. Atlikumu (Residuals) sadalījums
residuals = y_eng - y_pred
sns.histplot(residuals, kde=True, color='purple', bins=50, ax=axes[1])
axes[1].axvline(x=0, color='r', linestyle='--', lw=2)
axes[1].set_title('Atlikumu (Residuals) sadalījums')
axes[1].set_xlabel('Atlikums (Faktiskā - Prognozētā)')
axes[1].set_ylabel('Biežums')
axes[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
